# Basic Program for Deep Learning


**Import libraries**

Import the tensorflow and numpy.
Also import the framework to develop a Neural Network i.e. keras

```
import tensorflow as tf
import numpy as np

from tensorflow import keras
```

**Define and compile the neural network**

1. Create a simple neural net with 1 layer with 1 neuron. 
2. Compile the neural net
      For compiling a neural net, specify "loss" function and optimizer
      
      Loss function : funtion to measure the error in hypothesis. e.g. mean squared error (MSE)
      Optimizer : used to make another guess. e.g. Stochastic gradient descent (SGD), RMS Prop 

```
model = keras.models.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

**Providing Data**

For example provide 'x' values and corresponding 'y' values. 

```
xs = np.array([i for i in range(1,10)])
ys = np.array([(2*i + 50 + i) for i in range(1,10)])
```

```
print(xs)
print(ys)
```

**Train the Network**

Training is when the neural net learns about the relation between input xs and output ys.

Following tasks are done in loop -
   1. makes a guess
   2. learns the error in prediction
   3. uses optimizer to make another guess


```
model.fit(xs, ys, epochs=1000)
```

**Prediction**

Here we predict the output for a random input.

```
print(model.predict([20]))
```
